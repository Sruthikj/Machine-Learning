```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/breast_cancer_parameters_dataset.csv')
```

```python
word_to_number = {'Yes':1, 'No':0}
```

```python
data['Alcohol'] = data['Alcohol'].map(word_to_number)
```

```python
data.head()
```

```python
data['Obesity'] = data['Obesity'].map(word_to_number)
data['Radiation_exposure_to_chest'] = data['Radiation_exposure_to_chest'].map(word_to_number)
data.head()
```

```python
data['Nulliparity'] = data['Nulliparity'].map(word_to_number)
data['No_breast_feeding'] = data['No_breast_feeding'].map(word_to_number)
data.head()
```

```python
data['Late_menopause'] = data['Late_menopause'].map(word_to_number)
data.head()
```

```python
data['Family_history'] = data['Family_history'].map(word_to_number)
data['Early_menarche'] = data['Early_menarche'].map(word_to_number)
data.head()
```

```python
data.shape
```

```python
x = data[['Family_history', 'Early_menarche', 'Late_menopause', 'Nulliparity', 'No_breast_feeding', 'Alcohol', 'Obesity', 'Radiation_exposure_to_chest']]
y = data['Cancer_diagnosis']
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
correct = 0
wrong = 0

for i in range(len(y_pred)):
  predicted = 1 if y_pred[i] >= 0.5 else 0
  if y_test.iloc[i] == predicted:
    correct += 1
  else:
    wrong += 1

print(f"Total : {len(y_test)} , Correct Prediction : {correct} , Wrong Prediction : {wrong}")
```

```python
accuracy = correct/len(y_test)
print(f"Accuracy: {accuracy:.0%}")
```
